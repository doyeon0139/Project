```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./술_특징모음_last.csv', index_col=0)
df
```

```python
stw = pd.read_csv('stopwords-ko.txt', header=None)       # 불용어 제외
stopwords = list(stw.iloc[:, 0].values)
stopwords

def remove_stopwords(doc):
    text = [w for w in doc.split() if w not in stopwords]
    return ' '.join(text)

df.contents = df.contents.apply(remove_stopwords)
```

```python
docs = df.contents
```

```python
docs
```

```python
from krwordrank.word import summarize_with_keywords

keywords = summarize_with_keywords(docs, 
                                   min_count=1, 
                                   max_length=10, 
#                                    beta=0.85, 
                                   max_iter=10, 
                                   stopwords=stopwords, 
                                   verbose=True)
keywords = summarize_with_keywords(docs)
```

```python
print(keywords)
```

```python
from krwordrank.word import KRWordRank

wordrank_extractor = KRWordRank(
    min_count = 1, # 단어의 최소 출현 빈도수 (그래프 생성 시)
    max_length = 10, # 단어의 최대 길이
    verbose = True
    )

# beta = 0.85    # PageRank의 decaying factor beta
# max_iter = 10

keywords, rank, graph = wordrank_extractor.extract(docs, beta, max_iter)
```

```python
for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:30]:
    print('%8s:\t%.4f' % (word, r))
```

```python
print(keywords)
```

```python
from krwordrank.hangle import normalize
 
min_count = 5   # 단어의 최소 출현 빈도수 (그래프 생성 시)
max_length = 10 # 단어의 최대 길이
verbose =True
wordrank_extractor = KRWordRank(min_count, max_length , verbose)
 
beta = 0.85    # PageRank의 decaying factor beta
max_iter = 10

 
texts = [normalize(text,english=False , number=True) for text in docs ]
keywords, rank, graph = wordrank_extractor.extract(texts, beta, max_iter)
 
for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:30]:
        print('%8s:\t%.4f' % (word, r))
```

```python
from krwordrank.sentence import summarize_with_sentences

stop = 0

for b_idx in docs[5:]:
    print(b_idx)
    stop += 1
    test = b_idx
    st = ''
    for r in range(len(test)):
        texts = test[r]
#         texts = preprocessing(texts, b_idx)
        st += texts
    texts = st.split('. ')
    try:
        stopwords = {b_idx.split(' ')[0], b_idx.split(' ')[1]}
        keywords, sents = summarize_with_sentences(
                                               texts, 
                                               stopwords = stopwords,
                                               num_keywords=100, 
                                               num_keysents=10
                                               )
    except ValueError:
        print('key가 없습니다.')
        print()
        continue
        
    for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:7]:
        #print('%8s:\t%.4f' % (word, r))
        print('#%s' % word)
    print()
    if stop == 50:
        break
```

```python
from krwordrank.sentence import summarize_with_sentences

name_li = []
keyword_li = []
stop = 0

for i,b_idx in enumerate(docs[:]):
    
    print(df['상품명'][i])
    name = df['상품명'][i]
    stop += 1
    test = b_idx
    st = ''
    word_li = []
    for r in range(len(test)):
        texts = test[r]
#         texts = preprocessing(texts, b_idx)
        st += texts
    texts = st.split('. ')
    try:
        stopwords = {b_idx.split(' ')[0], b_idx.split(' ')[1]}
        keywords, sents = summarize_with_sentences(
                                               texts, 
                                               stopwords = stopwords,
                                               num_keywords=100, 
                                               num_keysents=10
                                               )
    except ValueError:
        print('key가 없습니다.')
        name_li.append(name)
        keyword_li.append(np.nan)
        print()
        continue
        
    for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:7]:
        try:
            #print('%8s:\t%.4f' % (word, r))
            word_li.append('#'+word)
            print('#%s' % word)
        except:
            word_li.append(np.nan)
            
    print()
    name_li.append(name)
    keyword_li.append(word_li)
    
#     if stop == 50:
#         break
```
