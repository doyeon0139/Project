# 토큰화

```python
df = pd.read_csv('./술_Text.csv', index_col=0)
df
```

```python
df.iloc[93,3] = '포도를 증류한 브랜디에서 나는 향긋한 향기가 물씬 올라와 기분을 좋게 만들어 준다. 술을 입에 넣으면 마치 맑은 청주를 마시는 느낌이다. 도수가 높지만 매우 부드럽고 술을 마시고 난 후에 자극적이지 않아 누구나 편하게 마실 수 있는 술이다. 양념이 과하지 않은 음식과 잘 어울린다.'
```

```python
def text_clean(doc):                    # 정규표현식을 이용하여 한글이 아닌것을 제외
    text = re.sub('[^가-핳ㄱ-ㅎㅏ-ㅣ ]','', doc)  
    return text
```

```python
# df['상품명'] = df['상품명'].apply(lambda x : text_clean(x))
df['정보'] = df['정보'].apply(lambda x : text_clean(x))
df['특징'] = df['특징'].apply(lambda x : text_clean(x))
df['맛_음식'] = df['맛_음식'].apply(lambda x : text_clean(x))
```

```python
df
```

```python
stw = pd.read_csv('stopwords-ko.txt', header=None)       # 불용어 제외
stopwords = list(stw.iloc[:, 0].values)
stopwords

def remove_stopwords(doc):
    text = [w for w in doc.split() if w not in stopwords]
    return ' '.join(text)
```

```python
df['정보'] = df['정보'].apply(remove_stopwords)
df['특징'] = df['특징'].apply(remove_stopwords)
df['맛_음식'] = df['맛_음식'].apply(remove_stopwords)
```

```python
kor = Komoran()
```

```python
docs = []

for n in df['맛_음식']:
    docs.append(kor.nouns(n))
    
docs
```

```python
def remove_stopwords_a(doc):
    try:
        text = [w for w in doc if w not in stopwords]
        return ' '.join(text)
    except:
        pass
```

```python
remove_stopwords_a(docs)
```

```python
for i in range(len(docs)):
    docs[i] = ' '.join(docs[i])
```

```python
docs
```

```python
vectorizer = CountVectorizer()
```

```python
vectorizer.fit(docs)
```

```python
print(vectorizer.get_feature_names())
```

```python
print(vectorizer.vocabulary_)
```

```python
docs = vectorizer.transform(docs)
```
