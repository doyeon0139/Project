```python
from collections import defaultdict
a_list =[]
content_1 = []
content_2 = []
content_3 = []
content_4 = []
content_5 = []
content_6 = []

contents = [content_1, content_2, content_3, content_4, content_5, content_6]

for pages in range(1,38):
    for page in range(1,16):
        url = 'https://terms.naver.com/list.naver?cid=42726&categoryId=58635&page={}'.format(pages)
        elem = driver.get(url)

        driver.find_element_by_css_selector(f'#content > div.list_wrap > ul > li:nth-child({page}) > div.info_area > div.subject > strong > a:nth-child(1)').click()

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        path = '#size_ct'

        sell = soup.select(path)[0]

        info_list = ['상품명','주종','도수','용량','가격','원재료','제조사','대표자명','주소','연락처','온라인스토어']

        tmpdict= defaultdict(list)
        for p in sell.select('div.att_type > div > div.wr_tmp_profile > div > table > tbody > tr'):
            tmpdict[p.select('th')[0].text].append(p.select('td')[0].text.strip(' \n'))

        a_list.append(tmpdict)

        full = []
        textli = []
        flag = 0
        sell = soup.select('#size_ct')[0]



        for content in sell.find_all(['p','h3','h4']): 
            try:
                print(content['class'])
                if content['class'][0]== 'txt':
                    textli.append(content.text)
                elif content['class'][0] == 'stress':
                    full.append(textli)
                    textli = [content.text]

                elif content['class'][0]== 'num':
                    textli.append(content.text)
            except:
                print('yabal')
                pass    

        for i,f in enumerate(full):
            contents[i].append(f)
            
    time.sleep(1)
```
