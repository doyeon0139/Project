```python
import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import warnings
warnings.filterwarnings(action='ignore')

import re
import time
```

```python
from urllib.request import urlopen
```

```python
driver = webdriver.Chrome()
```

```python
for pages in range(10,38):
    for page in range(1,16):
        url = 'https://terms.naver.com/list.naver?cid=42726&categoryId=58635&page={}'.format(pages)
        elem = driver.get(url)

        driver.find_element_by_css_selector(f'#content > div.list_wrap > ul > li:nth-child({page}) > div.info_area > div.subject > strong > a:nth-child(1)').click()

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        path = '#size_ct'

        sell = soup.select(path)[0]
        
        name = soup.select('#content > div.section_wrap > div.headword_title > h2')[0].text
#         print(name)

        for num,img in enumerate(sell.find_all(['img'])):
#             print(img)
            imgurl = img['origin_src']
#             print(imgurl)
            with urlopen(imgurl) as f:
                with open('./img/'+name+'_'+str(num)+'.jpg','wb') as h:
                    img = f.read()
                    h.write(img)
        print(name+' / 다운로드 완료')
```
